```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

#### Question 2.1

Load the Writing desk.jpg image file using the imread function and display it
using the imshow function. Add a title to the figure using the title function.

```python
img_desk = cv2.imread('Writing_desk.jpg')
img_desk = cv2.cvtColor(img_desk, cv2.COLOR_BGR2RGB)
plt.title('Writing Desk')
plt.imshow(img_desk)
```

#### Question 2.2

Extract separate rectangular image regions containing the painting, vase, and statuettes, and display them using the imshow function with a title.
Hints:
1) You may find it useful to display the whole image using the image function to find object positions.
2) Portions of matrices can be extracted by using the following syntax: matrixName (subscriptsDimension1, subscriptsDimension2, subscriptsDimension3) where matrixName is the name of the matrix, and the subscripts give the coordinates in each dimension to keep.

```python
fig = plt.figure(figsize=(15,10))
painting = img_desk[25:350, 100:420]
plt.subplot(1,3,1)
plt.title('Painting')
plt.imshow(painting)

vase = img_desk[270:485,510:630]
plt.subplot(1,3,2)
plt.title('vase')
plt.imshow(vase)

statuettes = img_desk[390:500,110:250]
plt.subplot(1,3,3)
plt.title('Statuettes')
plt.imshow(statuettes)
```

#### Question 2.3

Extract the separate colour channels for the rectangular image region containing the statuettes and display them individually using the imshow function with a title for each colour.
Hint: The different colour channels can be accessed using the third dimension of the image matrix.

```python
red_ch = statuettes[:,:,0]
green_ch = statuettes[:,:,1]
blue_ch = statuettes[:,:,2]
fig = plt.figure(figsize=(15,10))

plt.subplot(1,3,1)
plt.title('Red Channel')
plt.imshow(red_ch)

plt.subplot(1,3,2)
plt.title('Green Channel')
plt.imshow(green_ch)

plt.subplot(1,3,3)
plt.title('Blue Channel')
plt.imshow(blue_ch)
```

#### Question 2.4 

Combine the red, green, and blue image channels for the rectangular image region containing the statuettes into a single greyscale image and display the image using the imshow function with a title. Do this without a colour map by creating a separate greyscale image. Also display the same image using the jet, and hsv colour maps individually by using the colormap function. Hint: You may find it useful to convert the colour channels into a different data type when combining them.

```python
statuettes_gray = cv2.cvtColor(statuettes, cv2.COLOR_RGB2GRAY)
statuettes_hsv = cv2.cvtColor(statuettes, cv2.COLOR_RGB2HSV)
statuettes_jet = cv2.applyColorMap(statuettes, cv2.COLORMAP_JET)

fig = plt.figure(figsize=(15,10))

plt.subplot(1,3,1)
plt.title('GRAY')
plt.imshow(statuettes_gray, cmap='gray')

plt.subplot(1,3,2)
plt.title('HSV')
plt.imshow(statuettes_hsv)

plt.subplot(1,3,3)
plt.title('JET')
plt.imshow(statuettes_jet)
```

#### Question 2.5

Create a new version of the painting where the colour channels are remapped so that what was red becomes blue, green becomes red, and blue becomes green, and display the resulting image using the imshow function with a title.

```python
statuettes_new = statuettes.copy()
statuettes_new[:,:,0] = statuettes[:,:,2]
statuettes_new[:,:,1] = statuettes[:,:,0]
statuettes_new[:,:,2] = statuettes[:,:,1]

plt.imshow(statuettes_new)
```
