# Q2.2 
Extract separate rectangular image regions containing the painting, vase, and stat-
 uettes, and display them using the imshow function with a title.
 1) You may find it useful to display the whole image using the image function to find object
 positions.
 2) Portions of matrices can be extracted by using the following syntax:

```python
painting = img_rgb[30:345, 110:425]
plt.figure(figsize=(20,20))
vase = img_rgb[275:480,510:625]
statuettes = img_rgb[390:490,110:250]

the_array = [painting,vase,statuettes]
name_array = ['Painting', 'Vase', 'Statuettes']

for i in range(3):
    plt.subplot(1,3,i+1)
    plt.title(name_array[i])
    plt.imshow(the_array[i])
```

# Question 2.3 
Extract the separate colour channels for the rectangular image region containing
the statuettes and display them individually using the imshow function with a title for each
colour.
Hint: The different colour channels can be accessed using the third dimension of the image
matrix.

```python
# b = statuettes[:,:,0]
# g = statuettes[:,:,1]
# r = statuettes[:,:,2] or easily can be used this code -->

b,g,r = cv2.split(statuettes)
plt.figure(figsize=(20,20))
the_array = np.array([b,g,r])
name_array = ['Blue Channel', 'Green Channel', 'Red Channel']


for i in range(3):
    plt.subplot(1,3,i+1)
    plt.title(name_array[i])
    plt.imshow(the_array[i])
```

# Question 2.4
Combine the red, green, and blue image channels for the rectangular image region
containing the statuettes into a single greyscale image and display the image using the imshow
function with a title. Do this without a colour map by creating a separate greyscale image.
Also display the same image using the jet, and hsv colour maps individually by using the
colormap function. Hint: You may find it useful to convert the colour channels into a
different data type when combining them.

```python
gray = cv2.cvtColor(statuettes, cv2.COLOR_RGB2GRAY)
hsv = cv2.cvtColor(statuettes, cv2.COLOR_RGB2HSV)
jet = cv2.applyColorMap(statuettes, cv2.COLORMAP_JET)

fig = plt.figure(figsize=(20,20))
the_array = [gray,hsv,jet]
name_array = ['GRAY Channel', 'HSV Channel', 'JET Channel']


for i in range(3):
    plt.subplot(1,3,i+1)
    plt.title(name_array[i])
    plt.imshow(the_array[i])
```
