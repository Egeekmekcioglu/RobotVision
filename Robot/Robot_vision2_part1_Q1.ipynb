```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image, ImageEnhance
```

```python
plt.figure(figsize=(20,20))
img = cv2.imread("C:/Users/xboys/Downloads/Image/Robot/pizza.jpg")
pizza1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(1,3,1)
plt.title('Pizza')
plt.imshow(pizza1)

pizza2= Image.open("C:/Users/xboys/Downloads/Image/Robot/pizza.jpg")
enhancer = ImageEnhance.Contrast(pizza2)
enhance_pizza = enhancer.enhance(factor=1.5) # factor setting adjust the contrast 
plt.subplot(1,3,2)
plt.title('Enhanced Pizza')
plt.imshow(enhance_pizza)
```

```python
img = cv2.imread("C:/Users/xboys/Downloads/Image/Robot/circle.jpg")
circle = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.title('circle')
plt.imshow(circle)
```

```python
pizza1 = cv2.resize(pizza1,(615,461))
circle_resize = cv2.resize(circle,(615,461))
combining_img = cv2.bitwise_and(pizza1, circle_resize)
plt.imshow(combining_img)
```

```python
filename = 'ekmekcioglu_assignment2_1.jpg'
cv2.imwrite(filename, combining_img)
```
